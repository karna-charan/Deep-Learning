```python
text = "karna"
tokens = list(text)

vocab = {ch: i for i, ch in enumerate(set(tokens))}

embeddings = [vocab[ch] for ch in tokens]

print("Tokens:", tokens)
print("Embeddings:", embeddings)
```

```python
import torch
import torch.nn as nn

text = input ('enter the text')
tokens = text.lower().split()
vocab = {t:i for i,t in enumerate(set(tokens))}
encoded = torch.tensor([vocab[t] for t in tokens])

embed = nn.Embedding(len(vocab), 4)
vectors = embed(encoded)

print(vectors)
```
